```python
#importing libraries

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data_set = pd.read_csv("/content/Breast Cancer.csv")
```

```python
#checking null values
data_set.isna().sum()
```

```python
#dropping the Null value column
ds = data_set.dropna(axis=1)
```

```python
#shape after dropping null column
ds.shape
```

```python
ds.describe()
```

```python
#checking the value counts of diagnosis column
ds['diagnosis'].value_counts()
```

```python
# Label encoding M and B in diagnosis to 1 and 0 respectively
labelencoder_y = LabelEncoder()
ds.iloc[:,1] = labelencoder_y.fit_transform(ds.iloc[:,1].values)
```

```python
#checking label encoding
ds.head()
```

```python
#visualizing the co relation of 10 different factors with diagnosis
plt.figure(figsize=(10,10))
sns.heatmap(ds.iloc[:,1:10].corr(), annot=True, fmt=".0%")
```

```python
#separating diagnosis from other factors
x= ds.drop(columns="diagnosis", axis = 1)
y= ds['diagnosis']
```

```python
print(x)
print(y)
```

```python
#splitting training and testing data
x_train, x_test, y_train, y_test =train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 0)
```

```python
#standardizing the data
x_train=StandardScaler().fit_transform(x_train)
```

```python

forest = RandomForestClassifier(random_state=0, criterion="entropy", n_estimators=10)
forest.fit(x_train,y_train)
```

```python
print('Random forest accuracy:',forest.score(x_train,y_train))
```
